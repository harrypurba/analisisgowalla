```python
import pandas as pd
from sklearn import preprocessing
import numbers
```

```python
data = pd.read_csv("Documents/TBD/Gowalla_totalCheckins.txt", header=None, delimiter="\t")
time = data[1]
tanggalDict = dict([])
for x in time:
    tanggal = (x.split("T")[0])
    if tanggal in tanggalDict:
        tanggalDict[tanggal] += 1
    else:
        tanggalDict[tanggal] = 1

print(tanggalDict)
```

```python
dataout = pd.DataFrame(list(tanggalDict.items()),columns=['Date', 'Count'])
dataout.to_csv('tanggal.txt', sep='\t', encoding='utf-8',index=False)
```
